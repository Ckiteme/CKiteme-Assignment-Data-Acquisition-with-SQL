```python
# Load the pandas library to read our csv file 
# from an external the source. 
# ---
#
import pandas as pd
```

```python
#Read our file
Trafficaccidents = pd.read_csv('traffic_accidents.csv')
Trafficaccidents.head()
%sql DROP TABLE IF EXISTS Trafficaccidents;
%sql PERSIST Trafficaccidents;
%sql SELECT * FROM Trafficaccidents LIMIT 5;
```

```python
# Display a list of 5 makes that had the highest number of crashes?
%%sql
SELECT MAKE, SUM(MODEL) FROM Trafficaccidents
GROUP BY MAKE
ORDER BY SUM(MODEL) DESC
LIMIT 5
```

```python
#Most personal vehicles involved in crashes were traveling towards which direction?
%%sql

SELECT VEHICLE_USE, TRAVEL_DIRECTION, COUNT(TRAVEL_DIRECTION) FROM Trafficaccidents
WHERE VEHICLE_USE = 'PERSONAL'
GROUP BY TRAVEL_DIRECTION
ORDER BY COUNT(TRAVEL_DIRECTION)
```

```python
#How many passengers were involved in personal cars were in accidents?
%%sql

SELECT VEHICLE_USE, SUM(NUM_PASSENGERS  ) FROM Trafficaccidents

WHERE VEHICLE_USE = 'PERSONAL'
```

```python
#How many passengers were involved in accidents while traveling in personal Sport Utility vehicles?
%%sql

SELECT VEHICLE_TYPE, SUM(NUM_PASSENGERS ) FROM Trafficaccidents

WHERE VEHICLE_TYPE = 'SPORT UTILITY VEHICLE (SUV)'
```

```python
# Compute the total number of Mercedez-Benz vehicles involved in accidents while being parked.
%%sql

SELECT MAKE, MANEUVER, SUM(MANEUVER) FROM Trafficaccidents

WHERE MAKE = 'MERCEDES-BENZ' AND MANEUVER = 'PARKED' OR 'LEAVING TRAFFIC LANE TO PARK'
```

```python
#Compute the total occupant count per vehicle count

%%sql
SELECT MAKE, SUM(OCCUPANT_CNT) FROM Trafficaccidents
GROUP BY MAKE
ORDER BY SUM(OCCUPANT_CNT) DESC
```

```python
#Sort the number of passengers in descending order by travel direction taking into account travel direction. Return travel direction and no. of passengers.
%%sql
SELECT TRAVEL_DIRECTION, SUM(NUM_PASSENGERS) FROM Trafficaccidents
GROUP BY TRAVEL_DIRECTION
ORDER BY SUM(NUM_PASSENGERS) DESC

```

```python
#Select the top 5 vehicle models involved in the accident with no defect, order them by make, and show their maneuver.
%%sql

SELECT MODEL,COUNT(MODEL), MAKE, MANEUVER FROM Trafficaccidents

WHERE VEHICLE_DEFECT = 'NONE'

GROUP BY MODEL, MANEUVER

ORDER BY COUNT(MAKE) DESC

LIMIT 20
```

```python
#Which ridesharing vehicles in an accident had the most number of passengers? Return model, and the number of passengers.
%%sql

SELECT  MODEL, SUM(NUM_PASSENGERS) FROM Trafficaccidents

WHERE VEHICLE_USE = 'RIDESHARE SERVICE'

GROUP BY MODEL

ORDER BY SUM(NUM_PASSENGERS) DESC
```
